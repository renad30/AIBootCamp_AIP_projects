```python
# Example: Bar plot
survived_counts = df['Survived'].value_counts()
labels = ['Did not survive', 'Survived']

plt.bar(labels, survived_counts)
plt.xlabel('Survival')
plt.ylabel('Count')
plt.title('Count of Passengers by Survival')
plt.show()
```

Dataset Description
- sepal length (cm)
- sepal width (cm)
- petal length (cm)
- petal width (cm)
- target: Setosa (0), VersiColour (1), and Virginica (2)

```python
from sklearn import datasets #to import a dataset
```

```python
iris = datasets.load_iris() #we can open the libarary ro know the name
df= pd.DataFrame(data= np.c_[iris['data'], iris['target']], # ".c_" concatenates the two numpy arrays along second axis (ie. columns)
                 columns= iris['feature_names'] + ['target'])
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
# examine the data
df.head()
```

```python
# see the shape of the data using the shape.
df.shape
```

```python
# Display basic information about the dataset
df.info()
```

```python
# Display summary statistics of the numerical columns
df.describe()
```

```python
# Display the number of unique values in each columndf

df.value_counts()
```

```python
# Count of missing values in each column
df.isna().sum()
```

```python
# Visualize the distribution using
#Basic plot
plt.plot( 'sepal length (cm)', 'sepal width (cm)', data=df, linestyle='none', marker='o')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title('Relationship between sepal length & width')
plt.show()
```

```python
#The relationship between 'sepal width (cm)', 'petal length (cm)',
plt.plot( 'sepal width (cm)','petal length (cm)', data=df, linestyle='none', marker='o')
plt.xlabel('sepal width (cm)')
plt.ylabel('petal length (cm)')
plt.title('Relationship between sepal width & petal length')
plt.show()
```

```python
plt.boxplot(df['sepal length (cm)'])
plt.ylabel('sepal length (cm)')
plt.title('Sepal Length Box plot')
```
